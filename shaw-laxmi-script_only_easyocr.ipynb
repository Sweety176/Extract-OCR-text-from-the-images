```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from pylab import rcParams
from IPython.display import Image
rcParams['figure.figsize']=8,16
```

```python
reader=easyocr.Reader(['en'])
```

```python
output=reader.readtext('/testimage2.jpeg')
```

```python
output  # 1st part is the bounding box co-ordinates, 2nd part is the 'text', 3rd part is the confidence #
```

```python
cord=output[-1][0]
```

```python
x_min,y_min=[min(idx) for idx in zip(*cord)]
```

```python
x_max,y_max=[max(idx) for idx in zip(*cord)]
```

```python
image=cv2.imread('/testimage2.jpeg')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```
